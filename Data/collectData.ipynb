```python
import pandas as pd
import requests
```

```python


tokens = pd.read_csv("token.csv")
contracts = tokens["ContractAddress"]

mapping = {}
for i in contracts:
    url = "https://api.ethplorer.io/getTopTokenHolders/"+i+"?apiKey=EK-cQTo5-YwMrjdS-9bY9u&limit=100"
    jsonData = requests.get(url).json()
    if "error" not in jsonData:
        tokenDistribution = 0
        for j in jsonData['holders']:
            tokenDistribution = tokenDistribution + j['share']
        mapping[i] = tokenDistribution
```

```python
eth = "ETH"
newData = pd.read_csv("data.csv").query("Chain==@eth").iloc[:,:2]
contracts = newData["Address"]
newData
```

```python
mapping = {}
for i in newData["Address"]:
    url = "https://api.ethplorer.io/getTopTokenHolders/"+i+"?apiKey=EK-cQTo5-YwMrjdS-9bY9u&limit=50"
    jsonData = requests.get(url).json()
    if "error" not in jsonData:
        topShare = 0
        for j in jsonData['holders']:
            topShare += j['share']
        contractShare = 0
        for j in jsonData['holders']:
            address = j["address"]
            checkContractURL = "https://api.ethplorer.io/getAddressInfo/"+address+"?apiKey=EK-cQTo5-YwMrjdS-9bY9u"
            result = requests.get(checkContractURL).json()
            if("contractInfo" in result):
                contractShare += j["share"]
            # Check if address is contract and add share of it to the contractShare
        contractRatio = contractShare / topShare
        mapping[i] = [contractRatio, topShare]
```

```python
dataFrame = pd.DataFrame()
dataFrame["Address"] = []
dataFrame["TopHolding"] = []
dataFrame["ContractRatio"] = []
for i in mapping.keys():
    val = mapping[i]
    temp = pd.DataFrame()
    temp["Address"] = [i]
    temp["TopHolding"] = [val[1]]
    temp['ContractRatio'] = [val[0]]
    dataFrame = dataFrame.append(temp)
```
